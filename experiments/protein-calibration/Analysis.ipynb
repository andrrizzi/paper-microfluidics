```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from lxml import etree
import pandas as pd
```

```python
def getXPathToScan(name, well):
    return "/MeasurementResultData/Section[@Name='" + name + "']/Data[@Cycle='1']/Well[@Pos='" + well + "']/*"

def splitWellID(well):
    return (ord(well[0]) - 64), (int)(well[1:])

def dataFromXML(filename, platename):
    # Parse XML file.
    root = etree.parse(filename)

    elements = root.xpath("/MeasurementResultData/Section[@Name='" + platename + "']/Data[@Cycle='1']/Well")
    wellIDs = [element.attrib['Pos'] for element in elements]
    
    # Extract wavelenght from first well

    elements = root.xpath(getXPathToScan(platename, wellIDs[0]))
    data = [(float(element.attrib['WL']), float(element.text)) for element in elements]
    
    # Extract wavelength
    waveLengths = np.array(data)[:,0]
    
    # Extract all wells we are interested in
    bookOfScans = {}
    
    for wellID in wellIDs:
        elements = root.xpath(getXPathToScan(platename, wellID))
        data = [float(element.text) for element in elements]
        bookOfScans[wellID] = (np.array(data))

    return wellIDs, waveLengths, bookOfScans

def dataFromXMLSingle(filename, platename):
    # Parse XML file.
    root = etree.parse(filename)

    elements = root.xpath("/MeasurementResultData/Section[@Name='" + platename + "']/Data[@Cycle='1']/Well")
    wellIDs = [element.attrib['Pos'] for element in elements]
        
    # Extract all wells we are interested in
    bookOfScans = {}
    
    for wellID in wellIDs:
        elements = root.xpath(getXPathToScan(platename, wellID))
        data = [float(element.text) for element in elements]
        bookOfScans[wellID] = data[0]

    return wellIDs, bookOfScans
```
